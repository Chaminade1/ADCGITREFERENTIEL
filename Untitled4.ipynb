```python

import pandas as pd
import matplotlib.pyplot as plt

# Création du dataframe
data = {'Nom': ['Alice', 'Bob', 'Charlie', 'David', 'Fayçal','leila','abdoul','askia','Gama','Gomez','Nasir','Canon'],
        'Âge': [25, 30, 22, 35,26,18,20,29,20,15,28,45],
        'Score': [85, 92, 78, 95,32,42,52,62,76,35,65,78]}

df = pd.DataFrame(data)

# Sauvegarde du dataframe dans un fichier CSV
df.to_csv('donnees.csv', index=False)



# Chargement du fichier CSV dans un dataframe
df = pd.read_csv('donnees.csv')

# Affichage des premières lignes du dataframe pour vérification
print(df.head())

# Diagramme statistique (histogramme)
df['Âge'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
plt.title('Histogramme des Âges')
plt.xlabel('Âge')
plt.ylabel('Fréquence')
plt.show()

# Graphique linéaire
df.plot(x='Nom', y='Score', kind='line', marker='o', color='green')
plt.title('Graphique Linéaire des Scores')
plt.xlabel('Nom')
plt.ylabel('Score')
plt.show()

# Graphique circulaire (camembert) des Scores
df.plot(y='Score', kind='pie', labels=df['Nom'], autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Graphique Circulaire des Scores')
plt.show()
```
